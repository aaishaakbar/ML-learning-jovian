```python
df=pd.read_csv('IRIS.csv')
```

```python
numeric_cols=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
x=df[numeric_cols]
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3, random_state=42)
model.fit(x)
```

```python
model.cluster_centers_
```

```python
x
```

```python
preds=model.predict(x)
preds
```

```python
model.inertia_ #total varience of all the individual cluster can be find by this
```

```python
model=KMeans(n_clusters=6, random_state=42)
model.fit(x)
```

```python
preds=model.predict(x)
preds
```

```python
sns.scatterplot(data=x, x='sepal_length', y='petal_length', hue=preds)
```

```python
model.inertia_
```

```python

options = range(2, 11)
inertias = []

for n_clusters in options:
    model = KMeans(n_clusters=n_clusters, random_state=42).fit(x)
    inertias.append(model.inertia_)

plt.title('No of clusters vs Inertia')
plt.plot(options, inertias, '-o')
plt.xlabel('No. of clusters (K)')
plt.ylabel('Inertia')
plt.show()

```

```python
#DBSCAN
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=1.1, min_samples=4).fit(x)
```

```python
model.core_sample_indices_
```

```python
model.labels_
```

```python
sns.scatterplot(data=x, x='sepal_length', y='petal_length', hue=model.labels_)
```

# **Dimensionality Reduction**
reduct the number of columns without affecting the major information in that case we can say that if a dataset has 200 column and we reduce it in 5 columns without affecting major information then we can run the model more faster than 200 columns. one of this method is **PCA** and **t-SNE**. PCA is linear and t-SNE is non-linear.

```python
numeric_cols
```

```python
from sklearn.decomposition import PCA
```

```python
?PCA
```

```python
pca=PCA(n_components=2).fit(df[numeric_cols])
```

```python
pca
```

```python
pca.components_
```

```python
transformed=pca.transform(df[numeric_cols])
```

```python
sns.scatterplot(x=transformed[:,0], y=transformed[:,1], hue=df['species'])
```

```python
from sklearn.manifold import TSNE
```

```python
tsne=TSNE(n_components=2)
```

```python
transformed=tsne.fit_transform(df[numeric_cols])
```

```python
sns.scatterplot(x=transformed[:,0], y=transformed[:,1], hue=df['species'])
```
